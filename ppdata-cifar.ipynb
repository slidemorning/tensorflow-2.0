```python
import os
import glob
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

#### glob dataset path

```python
data_paths = glob.glob('dataset/cifar/train/*png')
len(data_paths)
```

```python
os.listdir('dataset/cifar/train/')[0]
```

```python
path = data_paths[0]
path
```

#### visualize CIimage using tensorflow.io

```python
gfile = tf.io.read_file(path)
image = tf.io.decode_image(gfile)
image.shape
```

```python
plt.imshow(image)
plt.show()
```

```python
def read_image(path):
    gfile = tf.io.read_file(path)
    image = tf.io.decode_image(gfile)
    return image
```

```python
for i in range(8):
    plt.imshow(read_image(data_paths[i]))
    plt.show()
```

### Make Batch
* (batch, height, width, channel)
* batch   data size(shape) 

#### images in List

```python
image_lst = []
for i in range(8):
    image = read_image(data_paths[i])
    #image = cv2.resize((32, 32))
    image_lst.append(image)
len(image_lst)
```

#### Numpy
* np.array()

```python
batch = np.array(image_lst)
batch.shape, type(batch)
```

#### Tensor
* tf.convert_to_tensor()

```python
batch = tf.convert_to_tensor(image_lst)
batch.shape, type(batch)
```
